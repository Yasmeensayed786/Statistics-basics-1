# 1. What is a random variable in probability theory?
Ans. In probability theory, a random variable is a variable that takes on different values based on the outcome of a random experiment.

There are two main types:

1. Discrete random variable – takes on a finite or countable number of values.
Example: the outcome of rolling a die (1, 2, 3, 4, 5, or 6).


2. Continuous random variable – takes on an infinite number of possible values within a given range.
Example: the exact height of a person, which can be any real number within a range.

# 3. What is the difference between discrete and continuous distributions?
Ans. The main difference between discrete and continuous distributions lies in the type of values the random variable can take and how their probabilities are represented.

1. Discrete Distribution

Values: Countable (finite or countably infinite) — like 0, 1, 2, ...

Probability: Assigned to individual outcomes.

Uses: Probability Mass Function (PMF)

Example:

Rolling a die: P(X = 4) = 1/6

Binomial distribution, Poisson distribution

2. Continuous Distribution

Values: Uncountable, from an interval or range — like any real number between 0 and 1

Probability: Assigned over intervals (P(X = a) = 0 for any single point)

Uses: Probability Density Function (PDF)

Example:

Height of people: P(160 cm < X < 170 cm) = some value

Normal distribution, Exponential distribution

In Simple Terms:

Discrete = specific "jumps" between values (you can list them)

Continuous = smooth range of values (you measure them)

# 11. What are the characteristics of a normal distribution?
Ans. Key Characteristics of a Normal Distribution:

1. Bell-shaped curve:
Symmetrical around the mean.

Most of the data clusters around the center.

2. Mean = Median = Mode:

The peak of the curve is at the mean.

Since it's symmetric, all three measures of central tendency are the same.

3. Defined by two parameters:

Mean (μ): Center of the distribution.

Standard deviation (σ): Measures the spread; larger σ means a wider, flatter curve.

4. Empirical Rule (68-95-99.7 Rule):

About 68% of the data falls within ±1σ from the mean.

About 95% within ±2σ.

About 99.7% within ±3σ.

5. Asymptotic:

The tails approach, but never touch, the horizontal axis.

6. Continuous and smooth:

The curve has no gaps or jumps, and every value within the range has some probability.

# 13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
Ans. The Central Limit Theorem states that: When you take a large enough number of random samples from any population (regardless of its original distribution), the distribution of the sample means will tend to follow a normal distribution, as long as the sample size is sufficiently large.


1. Allows us to use normal distribution tools:
Even when the population isn't normal, we can use the normal distribution to make inferences about sample means (like confidence intervals and hypothesis testing).

2. Justifies many statistical methods:

It’s the reason why Z-tests, t-tests, and regression models work so reliably.

3. Foundation of inferential statistics:

We often can’t measure an entire population, so we take samples. The CLT helps us understand how sample statistics relate to population parameters.

4. Applies broadly and practically:
It’s used in quality control, polling, economics, medicine, and virtually every field that uses data.

# 17. What are point estimates and interval estimates in statistics?
Ans. 1. Point Estimate:

A point estimate gives a single best guess of a population parameter based on sample data.

Example:
If you survey 100 students and find the average study time is 3 hours, that 3 hours is your point estimate for the population mean.

Common point estimates:

Sample mean  estimates population mean

Sample proportion  estimates population proportion

Sample standard deviation  estimates population standard deviation

2. Interval Estimate (Confidence Interval):

An interval estimate gives a range of values within which the population parameter is likely to fall, along with a confidence level.

Example:
"We are 95% confident that the average study time is between 2.5 and 3.5 hours."

This is a confidence interval, and it gives more information than just a single number.

Formula (for mean, when population σ is known):


\bar{x} \pm Z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}

Where:

 = sample mean

 = Z-score based on confidence level (e.g., 1.96 for 95%)

 = population standard deviation

 = sample size

 # 18. What is the significance of confidence intervals in statistical analysis?
 Ans. 1. Quantifies Uncertainty:
Confidence intervals provide a range that reflects the uncertainty around a sample estimate.

2. Gives More Insight than a Point Estimate:

Unlike a single value, CIs show the possible spread of the true population parameter.

3. Supports Better Decision-Making:

Helps assess whether a value (like 0 for no effect) lies within a range, aiding practical and statistical conclusions.

4. Shows the Effect of Sample Size and Variability:

Wider intervals suggest less precision (due to smaller samples or more variation); narrower intervals suggest greater precision.

5. Complements Hypothesis Testing:

CIs help interpret the results of hypothesis tests and often give a clearer picture of effect size and direction.

# 21. What are the assumptions for applying the Central Limit Theorem?
Ans.The Central Limit Theorem (CLT) is a fundamental statistical concept, but it relies on several assumptions for it to hold true. Here are the key assumptions:

1. Independent Observations:
The sampled observations should be independent of each other. This typically means using random sampling or random assignment.

2. Sample Size:
The sample size should be sufficiently large. While there's no hard rule, a common guideline is:

n ≥ 30 for most distributions.

Larger sample sizes may be needed if the population distribution is highly skewed or has heavy tails.

3. Identically Distributed:
The data should come from the same distribution (i.e., identically distributed). This is especially important if you're averaging multiple variables.

4. Finite Mean and Variance:
The population from which the samples are drawn should have a finite mean and variance.

```python
#10
# Simulating CLT with exponential distribution
samples = 1000
sample_size = 30
means = []

for _ in range(samples):
    sample = np.random.exponential(scale=2, size=sample_size)
    means.append(np.mean(sample))

plt.hist(means, bins=50, color='dodgerblue', edgecolor='black', density=True)
plt.title("Central Limit Theorem - Sampling Means from Exponential Distribution")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
#15
# One-sample Z-test
sample = np.random.normal(50, 10, 100)
sample_mean = np.mean(sample)
pop_mean = 52
pop_std = 10
n = len(sample)

z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))

print("Z-statistic:", z)
print("P-value:", p_value)
```

```python
#16
from scipy.stats import norm

data = np.random.normal(100, 15, 200)
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

conf = 0.95
z = norm.ppf(1 - (1 - conf) / 2)
margin = z * (std / np.sqrt(n))
ci = (mean - margin, mean + margin)

print("95% Confidence Interval:", ci)
```

```python
#17
data = np.random.normal(60, 8, 100)
mean = np.mean(data)
std = np.std(data, ddof=1)
z = norm.ppf(0.975)
ci = (mean - z * (std / np.sqrt(len(data))), mean + z * (std / np.sqrt(len(data))))

print(f"Confidence Interval: {ci}")
print("Interpretation: We are 95% confident the true mean lies within this interval.")
```
