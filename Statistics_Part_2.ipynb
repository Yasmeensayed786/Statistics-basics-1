#1. What is hypothesis testing in statistics?

Hypothesis testing is a statistical method used to make decisions or inferences about a population parameter based on sample data. It involves formulating two hypotheses (null and alternative), calculating a test statistic, and using it to decide whether to reject the null hypothesis.

#2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

The null hypothesis (H₀) is a statement of no effect or no difference, assumed true until evidence suggests otherwise. The alternative hypothesis (H₁ or Ha) contradicts the null and represents what we aim to support. For example, H₀: μ = 50 vs. H₁: μ ≠ 50.

#3. What is the significance level in hypothesis testing, and why is it important?

The significance level (α) is the threshold for rejecting the null hypothesis, commonly set at 0.05 or 5%. It represents the probability of making a Type I error (rejecting a true null). It controls how strict the test is.

#4. What does a P-value represent in hypothesis testing?
A P-value is the probability of obtaining results as extreme as (or more extreme than) the observed results, assuming the null hypothesis is true. It measures the strength of evidence against H₀.

#5. How do you interpret the P-value in hypothesis testing?

If the P-value ≤ α, reject the null hypothesis — the result is statistically significant. If P-value > α, fail to reject H₀ — there isn’t enough evidence to support the alternative.

#6. What are Type I and Type II errors in hypothesis testing?

Type I Error: Rejecting a true null hypothesis (false positive).

Type II Error: Failing to reject a false null hypothesis (false negative).
α is the probability of Type I error; β is the probability of Type II error.

#7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

One-tailed test checks for effect in one direction (e.g., H₁: μ > 50).

Two-tailed test checks for any difference, regardless of direction (e.g., H₁: μ ≠ 50).

#8. What is the Z-test, and when is it used in hypothesis testing?

The Z-test is used when the population variance is known, and the sample size is large (n > 30). It compares the sample mean to the population mean using the Z-distribution.

#9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

Z = (X̄ - μ) / (σ / √n)
It represents how many standard deviations the sample mean (X̄) is from the population mean (μ). It helps determine how unusual the result is under H₀.

#10. What is the T-distribution, and when should it be used instead of the normal distribution?

The T-distribution is used when the sample size is small (n < 30) and the population standard deviation is unknown. It accounts for additional variability due to estimating σ from the sample.

#11. What is the difference between a Z-test and a T-test?

Z-test: Known population variance, large sample.

T-test: Unknown population variance, small sample.
T-test uses the T-distribution which has heavier tails to account for uncertainty.

#12. What is the T-test, and how is it used in hypothesis testing?

A T-test compares sample means to determine if there is a significant difference. Types include one-sample, independent two-sample, and paired sample T-tests. It is used when σ is unknown.

#13. What is the relationship between Z-test and T-test in hypothesis testing?

Both test hypotheses about means. As sample size increases, the T-distribution approaches the Z-distribution, making their results similar for large samples.

#14. What is a confidence interval, and how is it used to interpret statistical results?

A confidence interval provides a range of values within which the true population parameter is likely to fall. For example, a 95% CI means we’re 95% confident the true mean lies in that range.

#15. What is the margin of error, and how does it affect the confidence interval?

The margin of error is the maximum expected difference between the sample estimate and the true population value. Larger margins create wider intervals, indicating less precision.

#16. How is Bayes' Theorem used in statistics, and what is its significance?

Bayes' Theorem updates the probability of a hypothesis based on new evidence. It's fundamental in Bayesian statistics, helping revise beliefs with observed data.

#17. What is the Chi-square distribution, and when is it used?

The Chi-square distribution is used in categorical data analysis. It’s applied in tests of independence, homogeneity, and goodness-of-fit, especially for count data in frequency tables.

#18. What is the Chi-square goodness of fit test, and how is it applied?

It checks whether observed frequencies match expected frequencies for a categorical variable. If the Chi-square statistic is large and P-value < α, we reject the hypothesis that the data fits the expected distribution.

#19. What is the F-distribution, and when is it used in hypothesis testing?

The F-distribution is used to compare variances between two or more groups. It’s the basis of ANOVA and the F-test for testing equality of variances.

#20. What is an ANOVA test, and what are its assumptions?

ANOVA (Analysis of Variance) compares means of 3 or more groups. Assumptions:

Independence of observations

Normally distributed groups

Equal variances (homogeneity of variance)

#21. What are the different types of ANOVA tests?

One-Way ANOVA: One factor/grouping variable

Two-Way ANOVA: Two independent variables

Repeated Measures ANOVA: Same subjects measured multiple times

#22. What is the F-test, and how does it relate to hypothesis testing?

The F-test compares two variances or group means in ANOVA. It calculates the ratio of variances. A large F-value with a small P-value suggests significant differences among groups.

```python
#1. Perform a Z-test for comparing a sample mean to a known population mean
import numpy as np
from scipy.stats import norm

# Sample data
sample = np.array([49, 52, 48, 47, 50, 51, 53, 50])
sample_mean = np.mean(sample)
sample_size = len(sample)
pop_mean = 50
pop_std = 2  # known population standard deviation

# Z-test calculation
z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

print("Z-score:", z_score)
print("P-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
#4. Two-tailed Z-test with visualization
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Z-test setup
sample_mean = 98
pop_mean = 100
pop_std = 15
n = 36
z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

# Plotting decision regions
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)
plt.plot(x, y, label='Z distribution')

# Shade critical regions
alpha = 0.05
z_critical = norm.ppf(1 - alpha/2)
plt.axvline(-z_critical, color='red', linestyle='--', label='Critical values')
plt.axvline(z_critical, color='red', linestyle='--')
plt.axvline(z_score, color='green', label='Observed Z')

plt.fill_between(x, 0, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.2)
plt.title("Two-Tailed Z-test")
plt.legend()
plt.grid()
plt.show()

print("Z-score:", z_score)
```

```python
#5. Function to calculate and visualize Type I and Type II errors
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_errors(mu0, mu1, sigma, n, alpha):
    se = sigma / np.sqrt(n)
    z_alpha = norm.ppf(1 - alpha)

    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)
    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    critical_value = mu0 + z_alpha * se
    beta = norm.cdf((critical_value - mu1) / se)

    plt.plot(x, y0, label='H0 Distribution', color='blue')
    plt.plot(x, y1, label='H1 Distribution', color='green')
    plt.axvline(critical_value, color='red', linestyle='--', label='Critical Value')
    plt.fill_between(x, 0, y0, where=(x >= critical_value), color='red', alpha=0.3, label='Type I Error')
    plt.fill_between(x, 0, y1, where=(x < critical_value), color='orange', alpha=0.3, label='Type II Error')

    plt.title('Type I and Type II Errors')
    plt.legend()
    plt.grid()
    plt.show()
    print("Type II Error Probability (β):", beta)

plot_errors(mu0=100, mu1=102, sigma=15, n=36, alpha=0.05)
```

```python
#9. Confidence interval for a sample mean
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)
    margin = t_crit * std_err
    return mean - margin, mean + margin

sample = np.random.normal(100, 15, 50)
ci = confidence_interval(sample)
print(f"95% Confidence Interval: {ci}")
```

```python
#10 Calculate Margin of Error for a Confidence Level
import numpy as np
from scipy.stats import t

def margin_of_error(data, confidence=0.95):
    n = len(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)
    return t_crit * std_err

sample = np.random.normal(100, 10, 30)
me = margin_of_error(sample)
print("Margin of Error:", me)
```

```python
#11 Bayesian Inference using Bayes' Theorem
def bayes_theorem(P_A, P_B_given_A, P_B_given_not_A):
    P_not_A = 1 - P_A
    P_B = P_B_given_A * P_A + P_B_given_not_A * P_not_A
    P_A_given_B = (P_B_given_A * P_A) / P_B
    return P_A_given_B

# Example
P_A = 0.01  # disease
P_B_given_A = 0.99  # test positive if disease
P_B_given_not_A = 0.05  # false positive
result = bayes_theorem(P_A, P_B_given_A, P_B_given_not_A)
print("Posterior Probability:", result)
```

```python
#16 F-test for Comparing Variances
import numpy as np
from scipy.stats import f

group1 = np.random.normal(100, 10, 20)
group2 = np.random.normal(100, 15, 20)

f_stat = np.var(group1, ddof=1) / np.var(group2, ddof=1)
df1, df2 = len(group1)-1, len(group2)-1
p_value = 1 - f.cdf(f_stat, df1, df2)

print("F-statistic:", f_stat)
print("P-value:", p_value)
```

```python
#20 Two-way ANOVA with Visualization
import statsmodels.api as sm
from statsmodels.formula.api import ols
import pandas as pd
import seaborn as sns

df = pd.DataFrame({
    'Score': np.random.randint(60, 90, 60),
    'Treatment': np.repeat(['A', 'B'], 30),
    'Group': ['X']*15 + ['Y']*15 + ['X']*15 + ['Y']*15
})

model = ols('Score ~ C(Treatment) + C(Group) + C(Treatment):C(Group)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)

sns.boxplot(x='Group', y='Score', hue='Treatment', data=df)
plt.title("Two-way ANOVA Visualization")
plt.grid()
plt.show()
```
